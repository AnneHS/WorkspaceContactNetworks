```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
import powerlaw
```

```python
width=240
height=200
```

```python
N=1113 # number of particles
T=3479 # simulation time
```

```python
def boundaries(x,width,height):
    if x[0] < 0 or x[0] > width or x[1] < 0 or x[1] > height:
        return False
    else:
        return True
```

```python
waitingtimes = powerlaw.Power_Law(xmin=7,parameters=[1.9])
```

```python
flighttimes=powerlaw.Stretched_Exponential(xmin=10,parameters=[0.0816,0.6561])
```

```python
## Simulate separate tracks
## CRW interrupted by rests

directions=np.random.uniform(-np.pi,np.pi,size=N)
tracks=np.zeros((N,T,2))

for i in range(N):
    if i%100==0:
        print(i)
    
    tracks[i][0,:] = np.random.uniform(size=2)*np.array([width,height])

    k=0

    while k < T:

        waitingtime = int(waitingtimes.generate_random(1)[0]) #int(np.random.exponential(100))#
        if k > 0:
            location = tracks[i][k-1,:]
        else:
            location = tracks[i][0,:]
            
        if k + waitingtime >= T:
            waitingtime = T - k 

        for t in range(k, k+waitingtime):
            tracks[i][t,:] = location
            
        k += waitingtime
        
        # Explore
        flighttime = int(flighttimes.generate_random(1)[0])#int(np.random.exponential(30))
        if k + flighttime >= T:
            flighttime = T - k

        for t in range(k,k+flighttime):
            turnangle = np.random.vonmises(0, kappa=4)
            theta = directions[i] + turnangle
            steplength = 2.5 #10 
            x = tracks[i][t-1,:] + steplength * np.array([np.cos(theta), np.sin(theta)])
            if boundaries(x, width,height):
                    tracks[i][t,:] = x
                    directions[i] = theta
            else:
                tracks[i][t,:] = tracks[i][t-1,:]
                directions[i] = np.random.uniform(-np.pi,np.pi)
        k += flighttime
```

```python
plt.figure(figsize=(8,6))

for i in range(3):
    plt.plot(tracks[i][:,0],tracks[i][:,1],linewidth=1)

plt.xticks(fontsize=16)
plt.yticks(fontsize=16)
plt.tick_params(axis='both',length=10,width=1,which='major')
plt.tick_params(axis='both',length=4,width=1,which='minor')

plt.xlim([-10,width+10])
plt.ylim([-10,height+10])

plt.show()
```
